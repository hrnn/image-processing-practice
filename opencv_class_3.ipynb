```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
#black image
black_img = np.zeros((512,512,3), np.uint8)
cv2.line(black_img, (0,0), (512,512),(255,0,0), 5)
cv2_imshow(black_img)
```

```python
cv2.rectangle(black_img, (384,0), (510,510), (0,0,255), 3)
cv2_imshow(black_img)
```

```python
cv2.circle(black_img, (128,128),63, (255,0,0), -1)
cv2.circle(black_img, (128,128),83, (255,0,0), 1)
cv2_imshow(black_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(black_img, 'Hi dude', (10,500), font, 4, (255,0,255), 2)
cv2_imshow(black_img)
```
