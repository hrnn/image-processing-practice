```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/gdrive/My Drive/Colab Notebooks/opencv/bird-milan-black-and-white-raptor.jpeg', 0)
cv2_imshow(img)
```

```python
img.shape
```

```python
new_image = np.ones((256,256), np.uint8)
black = np.zeros((128,128), np.uint8)
white = 255 * np.ones((128, 128), np.uint8)
gray_1 = 100 * np.ones((128, 128), np.uint8)
gray_2 = 200 * np.ones((128, 128), np.uint8)

new_image[0:128,0:128] = black
new_image[0:128,128:256] = white
new_image[128:256,0:128] = gray_1
new_image[128:256,128:256] = gray_2
cv2_imshow(new_image)
cv2.imwrite('/gdrive/My Drive/Colab Notebooks/opencv/new_image.jpg', new_image)
```

```python
colored_image = cv2.imread('/gdrive/My Drive/Colab Notebooks/opencv/fruit-basket.jpeg')
print(colored_image.shape)
cv2_imshow(colored_image)
```

```python
gray_1 = cv2.cvtColor(colored_image, cv2.COLOR_BGR2GRAY)
print(gray_1.shape)
cv2_imshow(gray_1)
```

```python
ret, threshold_1 = cv2.threshold(gray_1, 150, 255, cv2.THRESH_BINARY)
print(threshold_1.shape)
cv2_imshow(threshold_1)
```

```python
ret, threshold_2 = cv2.threshold(gray_1, 150, 255, cv2.THRESH_BINARY_INV)
print(threshold_2.shape)
cv2_imshow(threshold_2)
```
